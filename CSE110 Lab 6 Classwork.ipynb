Task 8

```python
user = input().split(", ")
dict1 = {}
for i in user:
    key,value = i.split(":")
    dict1[key] = int(value)
sum,count = 0,0
for values in dict1.values():
    sum += values
    count += 1
print(f"Average is {sum//count}.")
```

```python
threshold = int(input())
exam_marks = {'Cierra Vega': 175, 'Alden Cantrell': 200, 'Kierra Gentry': 165, 'Pierre Cox': 190}
dict1 = {}
for key,value in exam_marks.items():
    if value >= threshold:
        dict1[key] = value
print(dict1)
```
