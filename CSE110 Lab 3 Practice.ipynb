Task 16

```python
quantity = int(input())
max = 0
min = 0
sum = 0
for i in range(quantity):
  num = int(input())
  sum += num
  if num > max:
    max = num
  if num < min:
    min = num
avg = sum/quantity

print("Maximum",max)
print("Minimum",min)
print("Average is",avg)
```
