```python
class Student:

  def __init__(self, name, cgpa, credit = 9, dept = "CSE"):
    self.name = name
    self.cgpa = cgpa
    self.credit = credit
    self.dept = dept
    self.merit = False
    self.need = False

  def checkScholarshipEligibility(self):
    if self.credit > 10 and self.cgpa >= 3.7:
      self.merit = True
      print(f"{self.name} is eligible for Merit-based scholarship.")
    elif self.credit > 10 and (3.5 <= self.cgpa < 3.7):
      self.need = True
      print(f"{self.name} is eligible for Need-based scholarship.")
    else:
      print(f"{self.name} is not eligible for scholarship.")

  def showDetails(self):
    print(f"Name: {self.name}")
    print(f"Department: {self.dept}")
    print(f"CGPA: {self.cgpa}")
    print(f"Number of Credits: {self.credit}")
    if self.merit == True:
      print("Scholarship Status: Merit-based scholarship")
    elif self.need == True:
      print("Scholarship Status: Need-based scholarship")
    else:
      print("Scholarship Status: No scholarship")

print('--------------------------')
std1 = Student("Alif", 3.99, 12)
print('--------------------------')
std1.checkScholarshipEligibility()
print('--------------------------')
std1.showDetails()
print('--------------------------')
std2 = Student("Mim", 3.4)
std3 = Student("Henry", 3.5, 15,"BBA")
print('--------------------------')
std2.checkScholarshipEligibility()
print('--------------------------')
std3.checkScholarshipEligibility()
print('--------------------------')
std2.showDetails()
print('--------------------------')
std3.showDetails()
print('--------------------------')
std4 = Student("Bob", 4.0, 6, "CSE")
print('--------------------------')
std4.checkScholarshipEligibility()
print('--------------------------')
std4.showDetails()
```

```python
class Department:

  def __init__(self, dept = 'ChE Department', sec = 5):
    self.dept = dept
    self.sec = sec
    self.avg = 0
    print(f"The {dept} has {sec} sections.")

  def add_students(self, *students):
    total = 0
    if len(students) == self.sec:
      for student in students:
        total += student
      self.avg = round((total/len(students)),2)
      print(f"The {self.dept} has an average of {self.avg} students in each section.")
    else:
      print(f"The {self.dept} doesn't have {len(students)} sections.")

  def  merge_Department(self, *instances):
    new_total = 0
    for instance in instances:
      new_total += instance.avg*instance.sec
      print(f"{instance.dept} is merged to {self.dept}.")
    self.avg = round(((self.avg*self.sec)+new_total)/self.sec,2)
    return (f"Now the {self.dept} has an average of {self.avg} students in each section.")

d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21,30,40,36,10,32,27,51,45,15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
class Shopidify:

  def __init__(self, name = 'Guest'):
    self.name = name
    if name != 'Guest':
      print(f"Welcome {name} to Shopidify ")
    else:
      print(f"Welcome to Shopidify")
    self.cart = {}
    self.history = {}
    self.trans_list = []

  def add_to_cart(self, *items):
    if type(items[0]) == list:
      items = items[0]
    else:
      items = list(items)
    if len(items) == 1:
      items += [1]
    for i in range(0, (len(items)), 2):
      if items[i] in self.cart:
        self.cart[items[i]] += items[i+1]
      else:
        self.cart[items[i]] = items[i+1]

  def display_cart(self):
    print(f"Items in the cart for {self.name}: ")
    for item in self.cart:
      print(f"- {item}: {self.cart[item]}x")

  def checkout(self):
    print(f"Checkout completed for {self.name}")
    self.history = self.cart.copy()
    self.trans_list.append(self.history)
    self.cart = {}

  def display_history(self):
    print(f"Purchase history for {self.name}:")
    for i in range(len(self.trans_list)):
      print(f"Transaction {i+1}:")
      for item in self.trans_list[i]:
        print(f"- {item}: {self.trans_list[i][item]}x")

# Test the Shopidify class
guest_account = Shopidify()
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account = Shopidify("John")
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan", 2)
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart()
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.add_to_cart(["Chocolate Chip Cookies", 3,"Goku Action Figure",2,"Dumbbells-5kg",2])
john_account.display_cart()
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan")
guest_account.display_cart()
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.checkout()
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.display_history()
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.checkout()
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.display_history()
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
```

```python
class Author:

  def __init__(self,name = None):
    self.name = name
    self.books = {}
    self.book_count = 0

  def addBook(self,book_name,genre):
    if self.name == None:
      print("A book can not be added without author name")
    else:
      if genre in self.books:
        self.books[genre].append(book_name)
      else:
        self.books[genre] = [book_name]
      self.book_count += 1

  def printDetail(self):
    print(f"Number of Book(s): {self.book_count}")
    print(f"Author Name: {self.name}")
    for genre in self.books:
      print(f"{genre}: {', '.join(self.books[genre])}")

  def setName(self, name):
    self.name = name

a1 = Author()
print("=================================")
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("=================================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")
```
