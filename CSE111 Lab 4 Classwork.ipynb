```python
class Student:
    def __init__(self, name, id, dept = 'CSE'):
        self.name = name
        self.id = id
        self.dept = dept
        self.hours = None

    def dailyEffort(self, hours):
        self.hours = hours

    def printDetails(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.id}")
        print(f"Department: {self.dept}")
        print(f"Daily Effort: {self.hours} hour(s)")
        if self.hours <= 2:
            print("Suggestion: Should give more effort!")
        elif self.hours <= 4:
            print("Suggestion: Keep up the good work!")
        else:
            print("Suggestion: Excellent! Now motivate others.")

harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
class Farmer:
    def __init__(self, name_id = None):
        if name_id == None:
            print("Welcome to your farm!")
        elif type(name_id) == str:
            print(f"Welcome to your farm, {name_id}!")
        else:
            print(f"Welcome to your farm. Your farm ID is {name_id}!")

        self.crops = []
        self.fishes = []

    def addCrops(self, *args):
        if len(args) == 0:
            print("No crop(s) added.")
        else:
            for crop in args:
                self.crops.append(crop)
            print(f"{len(args)} crop(s) added.")

    def addFishes(self, *args):
        if len(args) == 0:
            print("No fish added.")
        else:
            for fish in args:
                self.fishes.append(fish)
            print(f"{len(args)} fish(s) added.")

    def showGoods(self):
        if len(self.crops) == 0:
            print("You don't have any crop(s).")
        else:
            print(f"You have {len(self.crops)} crop(s):")
            crp = ",".join(self.crops)
            print(crp)

        if len(self.fishes) == 0:
            print("You don't have any fish(s).")
        else:
            print(f"You have {len(self.fishes)} fish(s):")
            fsh = ",".join(self.fishes)
            print(fsh)

f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute", "Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")
```

```python
class TaxiLagbe:
    def __init__(self, taxi_no, area):
        self.taxi_no = taxi_no
        self.area = area
        self.dict1 = {}

    def addPassenger(self, *args):
        if len(self.dict1) < 4:
            for name_fare in args:
                name, fare = name_fare.split("_")
                self.dict1[name] = int(fare)
                print(f"Dear {name}! Welcome to TaxiLagbe.")
        else:
            print("Taxi Full! No more passengers can be added.")

    def printDetails(self):
        print(f"Trip info for Taxi number: {self.taxi_no}")
        print(f"This taxi can only cover the {self.area} area.")
        print(f"Total passengers {len(self.dict1)}")
        print(f"Passenger lists:")
        name_list = []
        total = 0
        for key, value in self.dict1.items():
            name_list.append(key)
            total += value
        names = ", ".join(name_list)
        print(names)
        print(f"Total collected fare: {total} Taka")

taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200','Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115', 'Parker_215')
print('-------------------------------')
taxi2.printDetails()
```
