```python
def even_checker(num):
    if num % 2 == 0:
        print("Even!!")
    else:
        print("Odd!!")

even_checker(2)
```

Task 2

```python
def fibonacci_checker(length):
    num1 = 0
    num2 = 1
    print(num1, end = " ")
    while num2 <= length:
        num3 = num1+num2
        print(num2, end = " ")
        num1 = num2
        num2 = num3

fibonacci_checker(10)
```

```python
def foo_moo(num):
    if num % 2 == 0 and num % 3 == 0:
        return "FooMoo"
    elif num % 3 == 0:
        return "Moo"
    elif num % 2 == 0:
        return "Foo"
    else:
        return "Boo"

foo_moo(6)
```

```python
def function_name(string):
    upper = 0
    lower = 0
    for i in string:
        if i.isupper() == True:
            upper += 1
        elif i.islower() == True:
            lower += 1
    print(f"No. of Uppercase characters : {upper}")
    print(f"No. of Lowercase Characters: {lower}")

function_name('HaRRy PotteR')
```

```python
age = int(input())
salary = int(input())
job = input().lower()

def calculate_tax(age,salary,job):
    if age < 18 or job == 'president' or salary < 10000:
        return 0
    elif 10000 <= salary <= 20000:
        tax = salary*5/100
        return tax
    elif salary > 20000:
        tax = salary*10/100
        return tax

calculate_tax(age,salary,job)
```

```python
days = int(input())

def function_name(days):

    year = days//365
    print(f"{year} years,", end = " ")

    months = (days-(year*365))//30
    print(f"{months} months and", end = " ")

    rem_days = days-(year*365)-(months*30)
    print(f"{rem_days} days")

function_name(days)
```

Task 7

```python
def show_palindrome(num):
    string = ""
    for i in range(1,num):
        string += str(i)
    for i in range(num,0,-1):
        string += str(i)
    return string
print(show_palindrome(3))
```

Task 8

```python
def show_palindromic_triangle(num):
    for i in range(1, num+1):
        for j in range(num-i):
            print(" ",end="")
        print(show_palindrome(i))
show_palindromic_triangle(5)
```

```python
import math
def area_circumference_generator(radius):
    area = math.pi*(radius**2)
    circumference = 2*math.pi*radius
    return (area,circumference)

print(area_circumference_generator(1.5))
x,y = area_circumference_generator(1.5)
print(f"Area of the circle is {x} and circumference is {y}")
```

```python
sentence = input().lstrip('"').rstrip('"')
position = int(input())

def function_name(sent,pos):
    new = ""
    remove = ""
    for i in range(len(sent)):
        if i == 0 or i % pos != 0:
            new += sent[i]
        elif i % pos == 0:
            remove += sent[i]
    result = new + remove
    return result

print(function_name(sentence, position))
```

```python
order_items = input()
def shoppy(order,location = "Dhanmondi"):
    items = {'"Rice"':105,'"Potato"':20,'"Chicken"':250,'"Beef"':510,'"Oil"':85}
    total = 0
    list1 = order.lstrip("[").rstrip("]").split(", ")
    for i in list1:
        if i in items:
            total += items[i]
    if location == "Dhanmondi":
        total += 30
    else:
        total += 70
    return total

shoppy(order_items, "Mohakhali")
```
