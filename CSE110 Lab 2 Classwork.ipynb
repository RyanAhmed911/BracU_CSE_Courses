Task 12

```python
hour = int(input())

if 0 <= hour <= 23:
  if 4 <= hour <= 6:
    print("Breakfast")
  elif 12 <= hour <= 13:
    print("Lunch")
  elif 16 <= hour <= 17:
    print("Snacks")
  elif 19 <= hour <= 20:
    print("Dinner")
  else:
    print("Patience is a virtue")
else:
  print("Wrong time")
```

Task 14

```python
distance_meter = float(input())
time_seconds = float(input())

distance_km = distance_meter/1000
time_hour = time_seconds/3600

velocity = distance_km/time_hour
print(velocity,"km/h")

if velocity < 60:
  print("Too slow. Needs more changes.")
elif velocity > 90:
  print("Too fast. Only a few changes should suffice.")
else:
  print("Velocity is okay. The car is ready!")
```
