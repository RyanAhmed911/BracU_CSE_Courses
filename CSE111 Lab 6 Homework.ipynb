```python
#Write your code here for subtasks 1-4.

class Student:

  ID = 0

  def __init__(self, name, department, age, cgpa):
    self.name = name
    self.department = department
    self.age = age
    self.cgpa = cgpa
    Student.ID += 1

  def showDetails(self):
    print(f"ID: {self.ID}\nName: {self.name}\nDepartment: {self.department}\nAge: {self.age}\nCGPA: {self.cgpa}")

  @classmethod
  def from_String(cls, info):
    name, department, age, cgpa = info.split("-")
    obj = cls(name, department, age, cgpa)
    return obj

s1 = Student("Samin", "CSE", 21, 3.91)
s1.showDetails()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.showDetails()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.showDetails()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.showDetails()

# Write the answer of subtask 5 here
print("The differences between class variables and instance variables are, class variables can be accessed and modified without instantiation while instance variables cannot. Also, class variables can store values that are constant accross all instances while instance variables can store different values for each object created.")
# Write the answer of subtask 6 here
print("The differences between class methods and instance methods are, we can call class methods without doing instantiation while we can only call instance methods after doing instantiation of the class. Class methods can only access and modify class variables and not instance variable while instance variables can access and modify both.")
```

```python
class Passenger:

  count = 0

  def __init__(self, name):
    self.name = name
    Passenger.count += 1
    self.fare = 450

  def set_bag_weight(self, weight):
    if 21 < weight <= 50:
      self.fare += 50
    elif weight > 50:
      self.fare += 100

  def printDetail(self):
    print(f"Name: {self.name}\nBus Fare: {self.fare}")

print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
class NikeBangladesh:

  branch = []
  stock = {'Air Jordan':0, 'Cortez':0, 'Zoom Kobe':0}
  sold = 0

  def __init__(self, name):
    self.name = name
    self.local_stock = {'Air Jordan':0, 'Cortez':0, 'Zoom Kobe':0}
    self.local_sold = 0
    NikeBangladesh.branch.append(name)

  @classmethod
  def status(cls):
    print(f"Nike Bangladesh Status:\nBranches Opened: {cls.branch}\nCurrently Stocked\n{cls.stock}\nSold: {cls.sold}")

  def details(self):
    print(f"Nike {self.name} outlet:\nProducts Currently Stocked:\n{self.local_stock}\nSold: {self.local_sold}")

  def restockProducts(self, products):
    for key in products:
      NikeBangladesh.stock[key] += products[key]
      self.local_stock[key] += products[key]

  def productSold(self, products):
    for key in products:
      NikeBangladesh.stock[key] -= products[key]
      self.local_stock[key] -= products[key]
      NikeBangladesh.sold += products[key]
      self.local_sold += products[key]

print("xxxxxxxxxxxxxx1xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
dhaka = NikeBangladesh("Dhaka Banani")
chittagong = NikeBangladesh("Chittagong GEC")
print("xxxxxxxxxxxxxx2xxxxxxxxxxxxxxxx")
dhaka.details()
print("xxxxxxxxxxxxxx3xxxxxxxxxxxxxxxx")
chittagong.details()
print("xxxxxxxxxxxxxx4xxxxxxxxxxxxxxxx")
dhaka.restockProducts(
{"Air Jordan":1200,"Cortez":200,"Zoom Kobe":200})
chittagong.restockProducts(
{"Air Jordan":1000,"Cortez":250,"Zoom Kobe":100})
print("xxxxxxxxxxxxxx5xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
print("xxxxxxxxxxxxxx6xxxxxxxxxxxxxxxx")
dhaka.productSold({"Air Jordan":760,"Cortez":90})
chittagong.productSold({"Air Jordan":520,"Zoom Kobe":70})
print("xxxxxxxxxxxxxx7xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
```

```python
class SultansDine:

  branch = 0
  total_sell = 0
  branch_list = []

  def __init__(self, name):
    self.name = name
    self.branch_sell = 0
    SultansDine.branch += 1
    SultansDine.branch_list.append(self)

  @classmethod
  def details(cls):
    print(f"Total Number of branch(s): {cls.branch}\nTotal Sell: {cls.total_sell} Taka")
    for i in cls.branch_list:
      percentage = (i.branch_sell/cls.total_sell)*100
      print(f"Branch Name: {i.name}, Branch Sell: {i.branch_sell} Taka\nBranch consists of total sell's: {round(percentage,2)}%")

  def sellQuantity(self, quantity):
    if quantity < 10:
      self.branch_sell = quantity*300
    elif quantity < 20:
      self.branch_sell = quantity*350
    else:
      self.branch_sell = quantity*400
    SultansDine.total_sell += self.branch_sell

  def branchInformation(self):
    print(f"Branch Name: {self.name}\nBranch Sell: {self.branch_sell} Taka")

SultansDine.details()
print('########################')
dhanmondi = SultansDine('Dhanmondi')
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
