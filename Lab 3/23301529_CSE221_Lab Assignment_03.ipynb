Task 1

```python
def merge(a, b):
    i = 0
    j = 0
    c = []
    
    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            c.append(a[i])
            i += 1
        else:
            c.append(b[j])
            j += 1
    
    if i == len(a):
        c += b[j:]
    elif j == len(b):
        c += a[i:]
    return c

def mergeSort(arr):
    if len(arr) <= 1:
        return arr
    else:
        mid = len(arr)//2
        a1 = mergeSort(arr[0:mid])
        a2 = mergeSort(arr[mid:])
        return merge(a1, a2)

file = open("input1.txt","r")
data = file.readlines()
file.close()
file = open("output1.txt","w")

n = data[0]
list1 = list(map(int, data[1].split()))
sorted_list = mergeSort(list1)

for i in range(len(sorted_list)):
    file.write(f"{sorted_list[i]} ")

file.close()
```

```python
def max_finder(arr):
    if len(arr) <= 1:
        return arr[0]
    else:
        mid = len(arr)//2
        a = max_finder(arr[0:mid])
        b = max_finder(arr[mid:])
        if a > b:
            return a
        else:
            return b

file = open("input2.txt","r")
data = file.readlines()
file.close()
file = open("output2.txt","w")

n = data[0]
list1 = list(map(int, data[1].split()))
max_val = max_finder(list1)

file.write(f"{max_val}")
file.close()
```

```python
def taller_counter(arr):
    if len(arr) == 1:
        return arr, 0
    else:
        mid = len(arr)//2
        
        a = taller_counter(arr[0:mid])
        b = taller_counter(arr[mid:])
        c = merge_and_counter(a[0],b[0])
        
        sorted_arr = c[0]
        count = c[1]
        
        return sorted_arr, count + a[1] + b[1]
        
def merge_and_counter(a,b):
        i = 0
        j = 0
        counter = 0
        arr1 = []

        while i < len(a) and j < len(b):
            if b[j] < a[i]:
                counter += len(a) - i
                arr1.append(b[j])
                j += 1
            else:
                arr1.append(a[i])
                i += 1
            
        if i == len(a):
            arr1 += b[j:]
        else:
            arr1 += a[i:]
        
        return arr1,counter

file = open("input3.txt","r")
data = file.readlines()
file.close()
file = open("output3.txt","w")

n = data[0]
list1 = list(map(int,data[1].split()))
result = taller_counter(list1)
count = result[1]

file.write(str(count))
file.close()
```

```python
def max_val(arr):
    if len(arr) == 1:
        return arr[0], arr
    else:
        mid = len(arr)//2
        left = max_val(arr[0:mid])
        right = max_val(arr[mid:])
        max_left = max(left[1])
        min_sq = min(right[1])**2
        max_sq = max(right[1])**2
        if min_sq > max_sq:
            max_right = int((min_sq)**(1/2))
        else:
            max_right = int((max_sq)**(1/2))
        cross = max_left + (max_right**2)
        return max(left[0],right[0],cross), arr

file = open("input4.txt","r")
data = file.readlines()
file.close()
file = open("output4.txt","w")

n = data[0]
listy = list(map(int, data[1].split()))
result = max_val(listy)
file.write(str(result[0]))
file.close()
```

```python
def partition(arr, p, r):
    i = p-1
    pivot = arr[r]
    for j in range(p, r):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1],arr[r] = arr[r], arr[i+1]
    return i+1

def quickSort(arr, p, r):
    if p < r:
        q = partition(arr, p, r)
        quickSort(arr, p, q-1)
        quickSort(arr, q+1, r)

file = open("input5.txt","r")
data = file.readlines()
file.close()
file = open("output5.txt","w")

n = int(data[0])
list1 = list(map(int, data[1].split()))
quickSort(list1, 0, n-1)

for i in range(n):
    file.write(f"{list1[i]} ")

file.close()
```

```python
def partition(arr, p, r):
    i = p-1
    pivot = arr[r]
    for j in range(p, r):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1],arr[r] = arr[r], arr[i+1]
    return int(i+1)

def kth_smallest(arr, p, r, key):
    if p <= r:
        q = partition(arr, p, r)
        if q == key-1:
            return arr[q]
        elif key-1 < q:
            return kth_smallest(arr, p, q-1, key)
        else:
            return kth_smallest(arr, q+1, r, key)

file = open("input6.txt","r")
data = file.readlines()
file.close()
file = open("output6.txt","w")

list1 = list(map(int, data[1].split()))
n = int((data[0].split())[0])
m = int((data[2].split())[0])

for i in range(m):
    result = kth_smallest(list1, 0, n-1, int(data[3+i]))
    file.write(f"{result}\n")

file.close()
```
