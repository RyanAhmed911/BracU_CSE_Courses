```python
def hospital_fee(**fees):
  print(fees)
  max_amount = 0
  max_payer = []
  for key,value in fees.items():
    if value > max_amount:
      max_amount = value
      max_payer = [key]
    elif value == max_amount:
      max_payer.append(key)
  return max_amount, max_payer
max_amount, max_payer = hospital_fee(Mashrafe = 400, Bumrah = 900, Steyn = 1200, Cummins = 900, Wood = 400, Marsh = 700)
if len(max_payer) > 1:
    payer = ', '.join(max_payer)
else:
    payer = max_payer[0]
print(f"Highest fee was {max_amount} tk which was paid by {payer}.")
```

```python
def is_james_bond(list1):
  string = ""
  for i in range(len(list1)):
    if list1[i] == 7 or list1[i] == 0:
      string += str(list1[i])
  if string == "007":
    return True
  else:
    return False
is_james_bond( [1, 2, 4, 0, 0, 7, 5] )
```

```python
def assign_students_to_sections(sections, *student_names):
  dict1 = {}
  for section_name in sections:
    dict1[section_name] = []
  for name in student_names:
    value = 0
    for i in name:
      value += ord(i)
    section_index = value % len(sections)
    dict1[sections[section_index]] += [name]
  return dict1
print(assign_students_to_sections('ABCDE', 'Alice','Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace'))
```

```python
def username_generator(first_name, last_name, student_id, middle_name = ""):
  digit = str(student_id)
  first = first_name.upper()
  last = last_name.lower()
  if len(first) > 3:
    first = first[:3]
  if len(last) > 3:
    last = last[len(last)-3:]
  if len(digit) > 4:
    digit = digit[len(digit)-4:]
  username = first + middle_name + last + "_" + digit
  return username
first_name, middle_name, last_name, student_id = input("First Name:"), input("Middle Name:"), input("Last Name:"), int(input("Student ID:"))
print(username_generator(first_name, last_name, student_id))
print(username_generator(first_name, last_name, student_id, middle_name))
```

```python
def key_generator(*names):
  encrypt_list = []
  for name in names:
    key = name[0].lower()
    trim = name[1:len(name)-1]
    for i in range(-1,-len(trim)-1,-1):
      key += str(ord(trim[i]))
    key += name[-1].upper()
    encrypt_list.append(key)
  return encrypt_list
key_list = key_generator ("Alex", "Bob", "Trudy")
print("Encrypted Keys:", key_list)
```

```python
import random
def playRockPaperScissor(rounds):
  p1_score = 0
  com_score = 0
  for i in range(rounds):
    com = random.choice(['rock','paper','scissor'])
    p1 = input()
    print("Computer:",com)
    if p1 == 'rock':
      if com == 'scissor':
        p1_score += 1
      elif com == 'paper':
        com_score += 1
    elif p1 == 'paper':
      if com == 'rock':
        p1_score += 1
      elif com == 'scissor':
        com_score += 1
    elif p1 == 'scissor':
      if com == 'paper':
        p1_score += 1
      elif com == 'rock':
        com_score += 1
    if p1_score > com_score:
      winner = "You have won the game!"
    else:
      winner = "Computer has won the game!"
  return p1_score, com_score, winner
p1_score, com_score, winner = playRockPaperScissor(int(input()))
print(f"Your Score: {p1_score}\nComputer's Score: {com_score}\n{winner}")
```

```python
import random
def playRockPaperScissor(rounds):
  p1_score = 0
  com_score = 0
  for i in range(rounds):
    com = random.choice(['rock','paper','scissor'])
    p1 = input()
    print("Computer:",com)
    result = p1+com
    if result == 'rockscissor' or result == 'scissorpaper' or result == 'paperrock':
        p1_score += 1
    if result == 'scissorrock' or result == 'paperscissor' or result == 'rockpaper':
        com_score += 1
  if p1_score > com_score:
    winner = "You have won the game!"
  elif p1_score < com_score:
    winner = "Computer has won the game!"
  else:
    winner = "Draw!"
  return p1_score, com_score, winner
p1_score, com_score, winner = playRockPaperScissor(int(input()))
print(f"Your Score: {p1_score}\nComputer's Score: {com_score}\n{winner}")
```
