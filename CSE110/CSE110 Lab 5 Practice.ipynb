Task 20

```python
string = input().strip()

if string[0] == "'":
  org = string.strip("'")
  print("Original data:", org)

elif string[0] == '"':
  org = string.strip('"')
  print("Original data:", org)

org = org.strip()
nosqr = org.lstrip("[")
nosqr = nosqr.rstrip("]")
print("After removing square brackets:", nosqr)

strfrmt = nosqr.split(", ")
print("Numbers in string format with extra white spaces:", strfrmt)

for i in range(len(strfrmt)):
  strfrmt[i] = int(strfrmt[i].strip())
print("Final data (numbers in list format):", strfrmt)
```
