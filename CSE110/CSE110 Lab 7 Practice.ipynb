```python
def area_circumference_generator(radius):
  area = 3.141592653589793*(radius**2)
  circumference = 2*3.141592653589793*(radius)
  return (area,circumference)

print(area_circumference_generator(1))

area,circumference = area_circumference_generator(1)
print(f"Area of the circle is {area} and circumference is {circumference}")
```

```python
def make_square(parameter):
  start,end = parameter
  dict1 = {}
  for i in range(start,end+1):
    dict1[i] = i**2
  return dict1

make_square((5,9))
```

```python
def rem_duplicate(parameter):
  list1 = []
  for i in parameter:
    if i not in list1:
      list1.append(i)
  tup = tuple(list1)
  return tup

rem_duplicate(("Hi", 1, 2, 3, 3, "Hi",'a', 'a', [1,2]))
```

```python
def function_name(listy):
  new_list = []
  counter = 0
  remove = 0
  for i in listy:
    if i not in new_list:
      new_list.append(i)
      counter = 1
    else:
      counter += 1
      if counter < 3:
        new_list.append(i)
      else:
        remove += 1
  print("Removed:",remove)
  return new_list

function_name([1, 2, 3, 3, 3, 3, 4, 5, 8, 8, 3])
```

```python
operator = input()
first = float(input())
second = float(input())

def function_name(operatory,firsty,secondy):
  if operatory == "+":
    result = firsty + secondy
  elif operatory == "-":
    result = firsty - secondy
  elif operatory == "/":
    result = firsty / secondy
  elif operatory == "*":
    result = firsty * secondy
  return float(result)

function_name(operator, first, second)
```

```python
sentence = input().lstrip('"').rstrip('"')
position = int(input())

def function_name(rayb,positiony):
  new = ""
  removed = ""
  for i in range(len(rayb)):
    if i == 0 or i % positiony != 0:
      new += rayb[i]
    else:
      removed += rayb[i]
  final = new+removed
  return final

print(function_name(sentence,position))
```

```python
money = int(input())

def splitting_money(amount):
  result = ""
  note_list = [500,100,50,20,10,5,2,1]
  for i in note_list:
    notes = amount//i
    amount = amount - i*notes
    if notes > 0:
      result += (f"\n{i} Taka: {notes} note(s)")

  return result

print(splitting_money(money))
```

```python
def individul_bonus_calculation(name,earning,goal,bonus):
  extra = 0
  if goal > 30:
    extra = 10000
  elif 20 <= goal <= 30:
    extra = 5000
  total_bonus = goal*(bonus/100*earning)+extra
  print(f"{name} earned a bonus of {int(total_bonus)} Taka for {goal} goals.")

individul_bonus_calculation('Luis', 80000, 25, 10)
```
