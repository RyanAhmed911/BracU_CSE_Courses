Task 3

```python
num_list = []
for i in range(5):
    num = int(input())
    num_list.append(num)
print("Printing values from the list in reverse order:")
for i in range(len(num_list)-1,-1,-1):
    print(num_list[i])
```

Task 5

```python
given_list = ["hey", "there", "", "what's", "", "up", "", "?"]
modified_list = []
for i in range(len(given_list)):
    if given_list[i] != "":
        modified_list.append(given_list[i])
print("Original List:", given_list)
print("Modified List:", modified_list)
```

Task 9

```python
my_list = input().split(", ")
modified_list = []
for i in range(len(my_list)):
    my_list[i] = int(my_list[i])
print("Input list:", my_list)
for i in range(len(my_list)):
    if my_list[i] not in modified_list:
        modified_list.append(my_list[i])
print("Modified list:", modified_list)
```
