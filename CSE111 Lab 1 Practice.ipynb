String 1

```python
user = input()

capital = 0
small = 0
new = ""

for letter in user:
  if 65 <= ord(letter) <= 90:
    capital += 1
  elif 97 <= ord(letter) <= 122:
    small += 1

if capital > small:
  for letter in user:
    if 65 <= ord(letter) <= 90:
      new += letter
    elif 97 <= ord(letter) <= 122:
      new += chr(ord(letter)-32)
else:
  for letter in user:
    if 65 <= ord(letter) <= 90:
      new += chr(ord(letter)+32)
    elif 97 <= ord(letter) <= 122:
      new += letter

print(new)
```

String 2

```python
string = input()

is_number = False
is_word = False

for character in string:
  if 48 <= ord(character) <= 57:
    is_number = True
  if 65 <= ord(character) <= 90 or 97 <= ord(character) <= 122:
    is_word = True

if is_number == True and is_word == True:
  print("MIXED")
elif is_word == True and is_number == False:
  print("WORD")
elif is_number == True and is_word == False:
  print("NUMBER")
```

String 5

```python
user = input()

lower = False
upper = False
digit = False
special = False

for i in user:
  if 65 <= ord(i) <= 90:
    upper = True
  if 97 <= ord(i) <= 122:
    lower = True
  if 48 <= ord(i) <= 57:
    digit = True
  if i in '_$#@':
    special = True

missing = 0

if lower == False:
  print("Lowercase character missing", end = "")
  missing += 1
if upper == False:
  if missing > 0:
    print(", Uppercase character missing", end = "")
  else:
    print("Uppercase character missing", end = "")
  missing += 1
if digit == False:
  if missing > 0:
    print(", Digit missing", end = "")
  else:
    print("Digit missing", end = "")
  missing += 1
if special == False:
  if missing > 0:
    print(", Special character missing", end = "")
  else:
    print("Special character missing", end = "")

if (upper == True) and (lower == True) and (digit == True) and (special == True):
  print("OK")
```

List 1

```python
my_list = []
while True:
  user = input()
  if user == "STOP":
    break
  else:
    my_list.append(user)

distinct_list = []
for number in my_list:
  if number in distinct_list:
    continue
  else:
    distinct_list.append(number)

frequency = []
for number1 in distinct_list:
  count = 0
  for number2 in my_list:
    if number1 == number2:
      count += 1
  frequency.append(count)

for i in range(len(distinct_list)):
  print(f"{distinct_list[i]} - {frequency[i]} times")
```

List 2

```python
n = int(input())

max = 0
sum = 0
max_list = []

for i in range(n):
  lst = input().split(" ")
  lst = [int(j) for j in lst]
  for k in lst:
    sum += k
  if sum > max:
    max = sum
    max_list = lst
  sum = 0

print(max)
print(max_list)
```

List 4

```python
while True:
  user = input()

  if user == "STOP":
    break
  else:
    ub_jumper = True
    lst = user.split(" ")
    lst = [int(i) for i in lst]
    for j in range(len(lst)-1):
      if lst[j+1] > lst[j]:
        difference = lst[j+1] - lst[j]
      else:
        difference = lst[j] - lst[j+1]
      if 0 < difference < len(lst):
        continue
      else:
        ub_jumper = False
        break
    if ub_jumper == True:
      print("UB Jumper")
    else:
      print("Not UB Jumper")
```

Dictionary & Tuple 1

```python
lst1 = input().split(", ")
dict1 = {}
value_list = []

for i in lst1:
  key,value = i.split(": ")
  value = int(value)
  dict1[key] = value

lst2 = input().split(", ")
for i in lst2:
  key,value = i.split(": ")
  value = int(value)
  if key in dict1:
    dict1[key] += value
  else:
    dict1[key] = value

for i in dict1:
  if dict1[i] not in value_list:
    value_list.append(dict1[i])

for i in range(0, len(value_list)):
    for j in range(i+1, len(value_list)):
        if(value_list[i] > value_list[j]):
            temp = value_list[i]
            value_list[i] = value_list[j]
            value_list[j] = temp

print(dict1)
print("Values:", tuple(value_list))
```

Dictionary & Tuple 2

```python
dict1 = {}

while True:
  user = input()
  if user == "STOP":
    break
  else:
    if user not in dict1:
      dict1[user] = 1
    else:
      dict1[user] += 1

for i in dict1:
  print(f"{i} - {dict1[i]} times")
```

Dictionary & Tuple 3

```python
user = input().split(", ")
dict1 = {}
for i in user:
  key, value = i.split(" : ")
  dict1[key] = value

inv = {}
for key, value in dict1.items():
    if inv.get(value) == None:
      inv[value] = [key]
    else:
      inv[value] = inv.get(value) + [key]

print(inv)
```
