```python
class Student:

  def __init__(self, name, id, cgpa):
    self.name = name
    self.__id = id
    self.__cgpa = cgpa

  def get_id(self):
    return self.__id

  def get_cgpa(self):
    return self.__cgpa

  def setId(self, id):
    self.__id = id

class Department:

  def __init__(self, dept):
    self.dept = dept
    self.id_list = []
    self.instants = []
    self.student_count = 0

  def findStudent(self, id):
    if id in self.id_list:
      for i in self.instants:
        if id == i.get_id():
          print("Student info:")
          print(f"Student Name: {i.name}")
          print(f"ID: {i.get_id()}")
          print(f"CGPA: {i.get_cgpa()}")
    else:
      print("Student with this ID doesn't exist, Please give a valid ID")

  def addStudent(self, *student_class_list):
    for i in student_class_list:
      if i.get_id() in self.id_list:
        print("Student with the same ID already exists, Please try with another ID")
      else:
        self.id_list.append(i.get_id())
        self.instants.append(i)
        self.student_count += 1
        print(f"Welcome to {self.dept} department, {i.name}")

  def details(self):
    print(f"Department Name: {self.dept}")
    print(f"Number of student:{self.student_count}")
    print("Details of the students: ")
    for i in self.instants:
      print(f"Student name: {i.name}, ID: {i.get_id()}, cgpa: {i.get_cgpa()}")

s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1,s2,s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib",22301010,3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib",22201010,2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()
```

```python
class Spaceship:

  def __init__(self, name, capacity):
    self.name = name
    self.capacity = capacity
    self.load = 0
    self.item_list = []

  def load_cargo(self, item):
    if (self.load + item.get_weight()) > self.capacity:
      print(f"Warning: Unable to load {item.get_item()} inside {self.name}. Exceeds capacity by {self.load + item.get_weight() - self.capacity}.")
    else:
      self.load += item.get_weight()
      self.item_list.append(item.get_item())

  def display_details(self):
    print(f"Spaceship Name: {self.name}")
    print(f"Capacity: {self.capacity}")
    print(f"Current Cargo Weight: {self.load}")
    print(f"Cargo: {self.item_list}")

class Cargo:

  def __init__(self, name, weight):
    self.__item = name
    self.__weight = weight

  def get_item(self):
    return self.__item

  def get_weight(self):
    return self.__weight

# Creating spaceships
falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold)  # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium)  # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium)  # This should not exceed Enterprise's capacity
enterprise.display_details()
```

```python
class Student:

  def __init__(self, name, id, dept):
    self.name = name
    self.id = id
    self.dept = dept
    self.email = None
    self.password = None
    self.login_status = None
    print("Student object is created!")

class Usis:

  def __init__(self):
    print("USIS is ready to use!")
    self.advised_courses = []

  def login(self, name):
    if (name.email != None) and (name.password != None):
      name.login_status = "ON"
      print("Login successful!")
    else:
      print("Email and password need to be set.")

  def advising(self, name, *courses):
    if name.login_status == None:
      print("Please login to advise courses!")
    else:
      if len(courses) == 0:
        print("You haven't selected any courses.")
      elif len(courses) > 3:
        print("You need special approval to take more than 3 courses.")
      else:
        for course in courses:
          self.advised_courses.append(course)
        print("Advising successful!")

  def individualDetails(self, name):
    course_string = ", ".join(self.advised_courses)
    return f"Name: {name.name}\nID: {name.id}\nDepartment: {name.dept}\nAdvised courses: {course_string}"

rakib = Student("Rakib", 12301455, "CSE")
print("1***********************")
usis_obj = Usis()
print("2***********************")
usis_obj.login(rakib)
print("3***********************")
usis_obj.advising(rakib)
print("4***********************")
rakib.email = "rakib@hotmail.com"
rakib.password = "1234"
print("5***********************")
usis_obj.login(rakib)
print("6***********************")
usis_obj.advising(rakib)
print("7***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110", "CSE260")
print("8***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110")
print("9***********************")
print(usis_obj.individualDetails(rakib))
```

```python
class Train:

  def __init__(self, name, *stops):
    self.name = name
    self.stops = stops
    self.start = stops[0]
    self.destination = stops[-1]
    self.passenger_list = []
    print(f"Welcome aboard on {self.name}")
    print(f"Start: {self.start}")
    print(f"Destination: {self.destination}")

  def addPassenger(self, *passengers):
    for passenger in passengers:

      self.passenger_list.append(passenger)
      print(f"{passenger.name} welcome aboard")

      if len(passenger.stops) == 0:
        passenger.start = self.start
        passenger.destination = self.destination

      elif len(passenger.stops) == 1:
        passenger.start = passenger.stops[0]
        passenger.destination = self.destination

      elif len(passenger.stops) == 2:
        passenger.start = passenger.stops[0]
        passenger.destination = passenger.stops[-1]

      passenger.fair = 100 * abs(self.stops.index(passenger.start)-self.stops.index(passenger.destination))

  def allPassengerDetails(self):
    for passenger in self.passenger_list:
      print(f"Name: {passenger.name},Start: {passenger.start},Destination: {passenger.destination}, Fair: ${passenger.fair}")

class Passenger:

  def __init__(self, name, *stops):
    self.name = name
    self.stops = stops
    self.start = None
    self.destination = None
    self.fair = 0

t1 = Train('T1-Express','New York','Manhattan','Brooklyn','Boston')
print("1========================")
p1 =Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke","Manhattan")
p3 = Passenger("Hinata","Manhattan","Brooklyn")
print("2========================")
t1.addPassenger(p2,p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train('Europe-Express','London','Paris','Brussels','Turkey')
print("5========================")
p4 =Passenger("Max","London","Brussels")
p5 = Passenger("Eleven","Paris")
p6 = Passenger("Mike","Brussels")
t2.addPassenger(p4,p5,p6)
print("6========================")
t2.allPassengerDetails()
```

```python
class BracuStudent:
  def __init__(self, name, address):
    self.name = name
    self.home = address
    self.bus_pass = False

  def show_details(self):
    print(f"Student Name: {self.name}")
    print(f"Lives in {self.home}")
    print(f"Have Bus Pass? {self.bus_pass}")

  def get_pass(self):
    self.bus_pass = True

class BracuBus:
  def __init__(self, route, max = 2):
    self.route = route
    self.max = max
    self.passenger_count = 0
    self.passenger_list = []

  def show_details(self):
    print(f"Bus Route: {self.route}")
    print(f"Passengers Count: {self.passenger_count} (Max: {self.max})")
    print(f"Passengers On Board: {self.passenger_list}")

  def board(self, *passengers):
    if len(passengers) == 0:
      print("No passengers!")
    else:
      for passenger in passengers:
        if (passenger.bus_pass == True) and (passenger.home == self.route):
          if self.passenger_count == self.max:
            print("Bus is full!")
          else:
            self.passenger_count += 1
            self.passenger_list.append(passenger.name)
            print(f"{passenger.name} boarded the bus.")

        elif passenger.bus_pass != True:
          print("You don't have a bus pass!")

        elif passenger.home != self.route:
          print("You got on the wrong bus!")

st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()
```
