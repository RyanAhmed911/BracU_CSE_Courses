```python
class CoffeeMachine:

  def __init__(self, name):
    self.name = name
    self.ingredient_list = []

  def insertIngredients(self, *ingredient):
    for i in ingredient:
      self.ingredient_list.append(i)

  def getDetails(self):
    string = f"Brand Name: {self.name}\nIngredients: "
    string += ", ".join(self.ingredient_list)
    return string

cm1 = CoffeeMachine("Miyako")
cm1.insertIngredients("Coffee beans", "Milk", "Sugar")
print(cm1.getDetails())
```

```python
class Sphere:

  def __init__(self, id, radius = 1, color = 'White'):
    self.id = id
    self.radius = radius
    self.color = color
    self.volume = 0
    self.volume = (4/3)*3.1416*(self.radius**3)

  def printDetails(self):
    print(f"Sphere ID: {self.id}")
    print(f"Color: {self.color}")
    print(f"Volume: {self.volume}")

  def merge_sphere(self, *spheres):
    print("Spheres are being merged")
    for instance in spheres:
      self.volume += instance.volume
      if self.color != instance.color:
        self.color = "Mixed Color"

sphere1 = Sphere("Sphere 1")
print("1***************")
sphere1.printDetails()
print("2***************")
sphere2 = Sphere("Sphere 2", 3)
print("3***************")
sphere2.printDetails()
print("4***************")
sphere3 = Sphere("Sphere 3", 2)
print("5***************")
sphere3.printDetails()
print("6***************")
sphere3.merge_sphere(sphere1,sphere2)
print("7***************")
sphere3.printDetails()
print("8***************")
sphere4 = Sphere("Sphere 4", 5, "Purple")
print("9***************")
sphere4.merge_sphere(sphere3)
print("10***************")
sphere4.printDetails()
```

```python
class Spotify:

  def __init__(self, genre, *song):
    self.genre = genre
    self.song_list = list(song)

  def show_playlist(self):
    print(f"Genre: {self.genre}")
    song_list = ", ".join(self.song_list)
    return f"Song list: {(song_list)}"

rock= Spotify("Heavy Metal", "This Fire Burns", "Here I Am")
print(rock.show_playlist())
```

```python
class PizzaMachine:

  def __init__(self, name = 'Regular', size = 6):
    self.name = name
    self.size = size

  def customizePizza(self, toppings, spice = 'Regular'):
    if type(toppings) != list or len(toppings) == 0:
      return "No toppings specified! Can't bake pizza."
    else:
      spice_levels = ("Regular", "Hot", "Super Naga")
      if spice not in spice_levels:
        return "Sorry! Spice level not allowed. Can't bake pizza."
      else:
        top_str = ",".join(toppings)
        return f"Your {self.size}-inch {spice} spicy {self.name} Pizza is ready with {top_str} toppings. Enjoy!"

pizza1 = PizzaMachine()
order1 = pizza1.customizePizza(["Cheese", "Pepperoni"],"Hot")
print("1################################# ")
print(order1)
print("2================================")
pizza2 = PizzaMachine("Vege")
order2 = pizza2.customizePizza("Super Naga")
print("3#################################")
print(order2)
print("4================================")
pizza3 = PizzaMachine("Chicken Blast",12)
order3 = pizza3.customizePizza(["Mushroom"])
print("5#################################")
print(order3)
print("6================================")
pizza4 = PizzaMachine("Beef Bonanza",16)
order4 = pizza4.customizePizza(["Cheese","Beef kala bhuna"],"Mild")
print("7#################################")
print(order4)
print("8================================")
```

```python
class Exam:

  def __init__(self, exam_type, question, marks):
    self.exam_type = exam_type
    self.question = question
    self.marks = marks
    self.total = self.marks*self.question

  def detail(self):
    detail = f"Exam Type: {self.exam_type}\nNumber of questions: {self.question}\nMarks per questions: {self.marks}\nTotal Marks: {self.total}"
    return detail

e1 = Exam('Midterm', 2, 10)
print(e1.detail())
print("===========================")
e2 = Exam('Final', 3, 10)
print(e2.detail())
```

```python
class RickMote:

  def __init__(self):
    self.channels = [0,2,3,6,7,9]
    self.cbox_status = "OFF"
    self.current_channel = 0
    self.volume = 3

  def power(self):
    if self.cbox_status == "OFF":
      self.cbox_status = "ON"
    else:
      self.cbox_status = "OFF"

  def showInfo(self):
    print("ID Cable Box Status:")
    print(f"Cable Box is: {self.cbox_status}")
    if self.cbox_status == "ON":
      print(f"Channel:{self.current_channel}")
      print(f"Volume:{self.volume}")

  def changeChannel(self, input = None):
    if self.cbox_status == "ON":
      if input == None:
        for i in range(len(self.channels)):
          if self.channels[i] == self.current_channel:
            if (i+1) > (len(self.channels)-1):
              print("Can not change channel beyond 9!")
            else:
              self.current_channel = self.channels[i+1]
              break
      else:
        if input in self.channels:
          self.current_channel = input
        else:
          print("TV channel does not exist.")
    else:
      print("Power is turned off. Cannot change channel.")

  def changeVolumeLevel(self, input = None):
    if self.cbox_status == "ON":
      if input == None:
        self.volume += 1
      else:
        self.volume += input
    else:
      print("Power is turned off. Cannot change volume.")

oTV = RickMote()
oTV.power()
print("1.############################")
oTV.showInfo()
print("2.############################")
oTV.changeChannel()
oTV.changeVolumeLevel()
oTV.showInfo()
print("3.############################")
oTV.power()
oTV.showInfo()
print("4.############################")
oTV.power()
oTV.changeVolumeLevel(4)
oTV.changeChannel(3)
oTV.showInfo()
print("5.############################")
oTV.changeVolumeLevel(-2)
oTV.showInfo()
print("6.############################")
oTV.power()
oTV.changeChannel(9)
oTV.changeVolumeLevel(-1)
oTV.showInfo()
print("7.############################")
oTV.power()
oTV.changeChannel(11)
oTV.showInfo()
```
