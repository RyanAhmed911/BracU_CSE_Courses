String 1

```python
user = input()

upper = 0
lower = 0
new = ""

for i in user:
    if 65 <= ord(i) <= 90:
        upper += 1
    elif 97 <= ord(i) <= 122:
        lower += 1

if upper > lower:
    for i in user:
        if 65 <= ord(i) <= 90:
            new += i
        elif 97 <= ord(i) <= 122:
            new += chr(ord(i)-32)
else:
    for i in user:
        if 65 <= ord(i) <= 90:
            new += chr(ord(i)+32)
        elif 97 <= ord(i) <= 122:
            new += i

print(new)
```

String 2

```python
user = input()

is_number = False
is_word = False

for i in user:
    if 65 <= ord(i) <= 90 or 97 <= ord(i) <= 122:
        is_word = True
    elif 48 <= ord(i) <= 57:
        is_number = True

if is_number == True and is_word == False:
    print("NUMBER")
elif is_number == False and is_word == True:
    print("WORD")
else:
    print("MIXED")
```

String 3

```python
user = input()

start = 0
new = ""

for i in range(len(user)):
    if 65 <= ord(user[i]) <= 90:
        start = i
        break

for j in range(start+1,len(user)):
    if 65 <= ord(user[j]) <= 90:
        break
    else:
        new += user[j]

if len(new) > 0:
    print(new)
else:
    print("BLANK")
```

String 3*

```python
user = input()

start = ""
end = ""
new = ""

for i in range(len(user)):
    if 65 <= ord(user[i]) <= 90:
        if start == "":
            start = i
        else:
            end = i

new = user[start+1:end]

if len(new) > 0:
    print(new)
else:
    print("BLANK")
```

Task 4

```python
string1,string2 = input().split(", ")

common = ""

for i in string1:
    if i in string2:
        common += i

for i in string2:
    if i in string1:
        common += i

if len(common) > 0:
    print(common)
else:
    print("Nothing in common.")
```

List 1

```python
my_list = []
while True:
    user = input()
    if user == "STOP":
        break
    else:
        my_list.append(user)
distinct_list = []
for i in my_list:
    if i not in distinct_list:
        distinct_list.append(i)
for i in distinct_list:
    count = 0
    for j in my_list:
        if i == j:
            count += 1
    print(f"{i} - {count} times")
```
