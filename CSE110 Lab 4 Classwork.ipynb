```python
string = input()
index = int(input())
reverse_string = ""
for i in range(index,-1,-1):
    reverse_string += string[i]
result = reverse_string + string[index+1:]
print(result)
```

Task 3

```python
string = input()
is_binary = True
for i in string:
    if i != "0" and i != "1":
        is_binary = False
if is_binary == True:
    print("Binary Number")
else:
    print("Not a Binary Number")
```

```python
string = input()
for i in string[1::2]:
    if 97<=ord(i)<=122:
        print(chr(ord(i)-32),end="")
    else:
        print(i,end="")
```
