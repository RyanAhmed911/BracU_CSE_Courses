```python
import numpy as np
import heapq as hq
```

A* Search (Part 1)

```python
file = open("input1.txt","r")
data = file.readlines()
file.close()

n, m = map(int, data[0].split())
a, b = map(int, data[1].split())
c, d = map(int, data[2].split())
start = (a, b)
end = (c, d)

maze = np.zeros((n,m), dtype=str)

for i in range(n):
    for j in range(m):
        maze[i][j] = data[i+3][j]

heuristic_values = {}
for row in range(n):
    for col in range(m):
        if maze[row][col] == "0":
            heuristic_values[(row, col)] = abs(row - c) + abs(col - d)

priority_queue = []
visited = []

hq.heappush(priority_queue, (0 + heuristic_values[start], 0, start, ""))
result_path = ""
result_cost = 0

while priority_queue:
    fn, gn, popped_index, path = hq.heappop(priority_queue)

    if popped_index in visited:
        continue
    else:
        visited.append(popped_index)

    if popped_index == end:
        result_path = path
        result_cost = gn
        break

    if popped_index[0] + 1 < n and maze[popped_index[0] + 1][popped_index[1]] == "0":
        hq.heappush(priority_queue, (gn + heuristic_values[(popped_index[0] + 1, popped_index[1])], gn + 1, (popped_index[0] + 1, popped_index[1]), path+"D"))

    if popped_index[0] - 1 > -1 and maze[popped_index[0] - 1][popped_index[1]] == "0":
        hq.heappush(priority_queue, (gn + heuristic_values[(popped_index[0] - 1, popped_index[1])], gn + 1, (popped_index[0] - 1, popped_index[1]), path+"U"))

    if popped_index[1] + 1 < m and maze[popped_index[0]][popped_index[1] + 1] == "0":
        hq.heappush(priority_queue, (gn + heuristic_values[(popped_index[0], popped_index[1] + 1)], gn + 1, (popped_index[0], popped_index[1] + 1), path+"R"))

    if popped_index[1] - 1 > -1 and maze[popped_index[0]][popped_index[1] - 1] == "0":
        hq.heappush(priority_queue, (gn + heuristic_values[(popped_index[0], popped_index[1] - 1)], gn + 1, (popped_index[0], popped_index[1] - 1), path+"L"))

if result_path == "":
    print("-1")
else:
    print(result_cost)
    print(result_path)
```

A* Search (Part 2)

```python
file = open("input2.txt", "r")
data = file.readlines()
file.close()
n, m = map(int, data[0].split())
a, b = map(int, data[1].split())

heuristic_values = {}
graph = {}
is_admissible = True
inadmissible_nodes = []

for i in range(n):
    heuristic_values[int(data[2+i].split()[0])] = int(data[2+i].split()[1])

for i in range(m):
    if int(data[2+n+i].split()[0]) not in graph:
        graph[int(data[2+n+i].split()[0])] = []
    graph[int(data[2+n+i].split()[0])].append(int(data[2+n+i].split()[1]))
    if int(data[2+n+i].split()[1]) not in graph:
        graph[int(data[2+n+i].split()[1])] = []
    graph[int(data[2+n+i].split()[1])].append(int(data[2+n+i].split()[0]))

queue = []
visited = []
actual_values = {}

queue.append(b)
visited.append(b)
actual_values[b] = 0

while queue:
    popped_node = queue.pop(0)
    if popped_node == a:
        continue

    for node in graph[popped_node]:
        if node not in visited:
            visited.append(node)
            queue.append(node)
            actual_values[node] = actual_values[popped_node] + 1

for node in graph:
    if heuristic_values[node] > actual_values[node]:
        is_admissible = False
        inadmissible_nodes.append(node)

if is_admissible:
    print("1")
else:
    print("0")
    print(f"Here nodes {inadmissible_nodes} are inadmissible")
```
