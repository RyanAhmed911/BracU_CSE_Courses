```python
import math
import random
```

```python
current_player = int(input("Enter starting player (0 for Scorpion, 1 for Sub-Zero): "))
depth = int(input("Enter the depth of the game tree: "))
branching_factor = int(input("Enter the branching factor of the game tree: "))
```

```python
def utility_score_generator(depth, branching_factor):
    utility_score = {}
    for i in range(branching_factor ** depth):
        utility_score[i] = random.choice([-1, 1])
    return utility_score
```

```python
def minimax_with_alpha_beta_pruning(node_index, depth, is_maximizing, utility_score, alpha=-math.inf, beta=math.inf):
    if depth == 0:
        return utility_score[node_index]
    if is_maximizing:
        local_value = -math.inf
        for i in range(branching_factor):
            value = minimax_with_alpha_beta_pruning(node_index * branching_factor + i, depth - 1, False, utility_score, alpha, beta)
            local_value = max(local_value, value)
            alpha = max(alpha, local_value)
            if beta <= alpha:
                break
        return local_value
    else:
        local_value = math.inf
        for i in range(branching_factor):
            value = minimax_with_alpha_beta_pruning(node_index * branching_factor + i, depth - 1, True, utility_score, alpha, beta)
            local_value = min(local_value, value)
            beta = min(beta, local_value)
            if beta <= alpha:
                break
        return local_value
```

```python
rounds_played = 0
scorpion_wins = 0
sub_zero_wins = 0
round_winners = []

while scorpion_wins < 2 and sub_zero_wins < 2:
    utility_scores = utility_score_generator(depth, branching_factor)
    
    if current_player == 1:
        is_maximizing = True
    else:
        is_maximizing = False

    winner_score = minimax_with_alpha_beta_pruning(0, depth, is_maximizing, utility_scores, float('-inf'), float('inf'))
    
    if winner_score == -1:
        round_winners.append("Scorpion")
        scorpion_wins += 1
    else:
        round_winners.append("Sub-Zero")
        sub_zero_wins += 1
        
    rounds_played += 1
    current_player = 1 - current_player

if scorpion_wins == 2:
    game_winner = "Scorpion"
else:
    game_winner = "Sub-Zero"
    
print(f"Game Winner: {game_winner}")
print(f"Total Rounds Played: {rounds_played}")
for i, winner in enumerate(round_winners):
    print(f"Winner of Round {i + 1}: {winner}")
```
