```python
def BMI(height,weight):
    height = height/100
    bmi = round(weight/(height**2),1)
    condition = "Normal"
    if bmi < 18.5:
        condition = "Underweight"
    elif 25<=bmi<=30:
        condition = "Overweight"
    elif bmi > 30:
        condition = "Obese"
    print(f"Score is {bmi}. You are {condition}")
BMI(152,48)
```

```python
def foodpanda(burger,location = "Mohakhali"):
    menu = {'BBQ Chicken Cheese Burger':250,'Beef Burger':170,'Naga Drums':200}
    meal = 0
    for key,value in menu.items():
        if key == burger:
            meal += menu[key]
    tax = meal*0.08
    if location == "Mohakhali":
        total = meal + tax + 40
    else:
        total = meal + tax + 60
    return total
print(foodpanda('Beef Burger','Dhanmondi'))
```

```python
def replace_domain(email,new_dom,old_dom = "sheba.xyz"):
    name = ""
    final = ""
    for i in email:
        if i != '@':
            name += i
        else:
            name += i
            break
    if old_dom != "sheba.xyz":
        final += name + new_dom
        print(f"Changed: {final}")
    else:
        print(f"Unchanged: {email}")
replace_domain('alice@kaaj.com','sheba.xyz','kaaj.com')
replace_domain('bob@sheba.xyz','sheba.xyz')
```

```python
def palindrome(string):
    mod_str = ""
    reverse = ""
    for i in string:
        if i == " ":
            continue
        else:
            mod_str += i
    for i in range(-1,-len(mod_str)-1,-1):
        reverse += mod_str[i]
    if mod_str == reverse:
        print("Palindrome")
    else:
        print("Not a palindrome")
palindrome('nurses run')
```

```python
def calender(days):
    years = days//365
    rem_days = days - (years*365)
    months = rem_days//30
    rem_days -= (months*30)
    print(f"{years} years, {months} months and {rem_days} days")
calender(4000)
```

```python
def capitalization(string):
    final = ""
    final += chr(ord(string[0])-32)+string[1]
    for i in range(2,len(string)):
        if string[i-2] in '.!?':
            final += chr(ord(string[i])-32)
        elif string[i-1] == " " and string[i+1] == " " and string[i] == "i":
            final += chr(ord(string[i])-32)
        else:
            final += string[i]
    return final
print(capitalization("my favourite animal is a dog. a dog has sharp teeth so that it can eat flesh very easily. do you know my pet dog's name? i love my pet very much."))
```
