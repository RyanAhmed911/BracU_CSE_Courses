```python
string = input()
counter=0
for i in string:
  if 65 <= ord(i) <= 90 or 97 <= ord(i) <= 122:
    if counter%2 == 0:
      if 97 <= ord(i) <= 122:
        print(chr(ord(i)-32),end="")
        counter += 1
      elif 65 <= ord(i) <= 90:
       print(i,end="")
       counter += 1
    elif counter%2 != 0:
      if 65 <= ord(i) <= 90:
        print(chr(ord(i)+32),end="")
        counter += 1
      elif 97 <= ord(i) <= 122:
       print(i,end="")
       counter += 1
  else:
    print(i,end="")
```

```python
matches = int(input())
win = 0
points = 0

for i in range(matches):
  score = input()

  if int(score[0]) > int(score[2]):
    win += 1
    points += 3

  elif int(score[0]) == int(score[2]):
    points += 1

print("Total Points:", points)
print(f"Winning Percentage: {round(win/matches*100, 2)}%")
```
