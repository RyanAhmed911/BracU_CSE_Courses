Task 1 (A)

```python
def adjacency_matrix_generator(num_vertices, ve):
    arr = [[0]*num_vertices for k in range(num_vertices)]
    for i in range(len(ve)):
        arr[ve[i][0]][ve[i][1]] = ve[i][2]
    return arr

file = open("input1a.txt", "r")
data = file.readlines()
file.close()

num_vertices = int((data[0].split())[0]) + 1
ve = []
for i in range(1, len(data)):
    ve.append(list(map(int, data[i].split())))
result = adjacency_matrix_generator(num_vertices, ve)

file = open("output1a.txt", "w")
for i in range(num_vertices):
    for j in range(num_vertices):
        file.write(f"{result[i][j]} ")
    file.write("\n")
file.close()
```

```python
def unweighted_graph_generator(num_vertices, ve):
    dict1 =  {}
    for i in range(num_vertices):
        dict1[i] = []
    for j in range(len(ve)):
        dict1[ve[j][0]].append(ve[j][1])
        dict1[ve[j][1]].append(ve[j][0])
    return dict1

def shortest_path(graph, source, destination):
    parent = {}
    for v in graph:
        parent[v] = None
    visited = [source] 
    queue = [source]

    while queue != []:
        u = queue.pop(0)
        if u == destination:
            break
        else:
            for v in graph[u]:
                if v not in visited:
                    visited.append(v)
                    parent[v] = u
                    queue.append(v)
    path = []
    u = destination
    while u != None:
        path.append(u)
        u = parent[u]
    path.reverse()
    path_str = ""
    for i in range(len(path)):
        path_str += str(path[i]) + " "
    
    return f"Time: {len(path)-1}\nShortest Path: {path_str}"

file = open("input5.txt", "r")
data = file.readlines()
file.close()

num_vertices = int((data[0].split())[0]) + 1
ve = []
for i in range(1, len(data)):
    ve.append(list(map(int, data[i].split())))
destination = int((data[0].split())[2]) 

graph = unweighted_graph_generator(num_vertices, ve)
result = shortest_path(graph, 1, destination)

file = open("output5.txt", "w")
file.write(result)
file.close()
```

```python
def max_diamond_finder(matrix):
    movements = [(0, 1), (1, 0), (0, -1), (-1, 0)]

    def is_valid_move(x,y):
        return 0 <= x < row and 0 <= y < col and matrix[x][y] != '#'
    
    def bfs(i, j):
        diamond_count = 0
        visited = [[False]*col for k in range(row)]
        visited[i][j] = True
        queue = [(i, j)]
        while queue:
            u,v = queue.pop(0)
            if matrix[u][v] == "D":
                diamond_count += 1
            for dx, dy in movements:
                if is_valid_move(u+dx, v+dy) and not visited[u+dx][v+dy]:
                    visited[u+dx][v+dy] = True
                    queue.append((u+dx, v+dy))
        return diamond_count
    
    max_diamond = 0
    
    for i in range(row):
        for j in range(col):
            if matrix[i][j] == ".":
                max_diamond = max(max_diamond, bfs(i,j))
    return max_diamond

file = open("input6.txt", "r")
data = file.readlines()
file.close()

row = int((data[0].split())[0])
col = int((data[0].split())[1])

matrix = []
for i in range(1, row+1):
    r = []
    for j in range(col):
        r.append(data[i][j])
    matrix.append(r)

result = max_diamond_finder(matrix)

file = open("output6.txt", "w")
file.write(str(result))
file.close()
```

```python
def unweighted_undirected_graph_generator(num_vertices, pairs):
    adj_list = {}
    for i in range(num_vertices+1):
        adj_list[i] = []
    for i in range(len(pairs)):
        adj_list[pairs[i][0]].append(pairs[i][1])
        adj_list[pairs[i][1]].append(pairs[i][0])
    return adj_list

def bfs(graph, source):
    visited = [source]
    distance = [0] * (num_vertices+1)
    queue = [source]
    while queue:
        u = queue.pop(0)
        for v in graph[u]:
            if v not in visited:
                visited.append(v)
                distance[v] = distance[u] + 1
                queue.append(v)
    return distance

file = open("input7.txt","r")
data = file.readlines()
file.close()

num_vertices = int(data[0])
pairs = []
for i in range(1,len(data)):
    pairs.append(list(map(int,(data[i].split()))))

graph = unweighted_undirected_graph_generator(num_vertices, pairs)

distances1 = bfs(graph, 1)
start_max_distance = max(distances1)
starting_node = distances1.index(start_max_distance)

distances2 = bfs(graph, starting_node)
end_max_distance = max(distances2)
ending_node = distances2.index(end_max_distance)

file = open("output7.txt","w")
file.write(f"{starting_node} {ending_node}")
file.close()
```

```python
def unweighted_undirected_graph_generator(num_vertices, pairs):
    adj_list = {}
    for i in range(num_vertices+1):
        adj_list[i] = []
    for i in range(len(pairs)):
        adj_list[pairs[i][0]].append(pairs[i][1])
        adj_list[pairs[i][1]].append(pairs[i][0])
    return adj_list

def max_of_race(graph):
    colors = [0] * len(graph)
    for node in graph:
        colors[node] = -1
    colors[1] = 0
    queue = [1]

    while queue:
        u = queue.pop(0)
        for v in graph[u]:
            if colors[v] == -1:
                colors[v] = 1 - colors[u]
                queue.append(v)
    
    a = 0
    b = 0

    for i in range(len(colors)):
        if colors[i] == 0:
            a += 1
        elif colors[i] == 1:
            b += 1
    
    return max(a, b)

file = open("input8.txt","r")
data = file.readlines()
file.close()

file = open("output8.txt","w")

num_cases = int(data[0])
start = 1
for i in range(num_cases):
    ender = None
    for j in range(start, int(data[start])+2):
        ender = j
        pairs = []
        for k in range(start+1, start+int(data[start])+1):
            pairs.append(list(map(int, data[k].split())))
        graph = unweighted_undirected_graph_generator(int(data[start])+1, pairs)
        result = max_of_race(graph)
    file.write(f"Case {i+1}: {result}\n")
    start = ender + 1

file.close()
```
