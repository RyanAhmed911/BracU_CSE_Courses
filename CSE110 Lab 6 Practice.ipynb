```python
#Given dictionaries
dict1 = {'Harry':15, 'Draco':8, 'Nevil':19}
dict2 = {'Ginie':18, 'Luna': 14}

#Creating an empty dictionary
marks = {}

#Putting the items of the dictionaries in the empty dictionary using a loop
for i in dict1:
  marks[i] = dict1[i]
for j in dict2:
  marks[j] = dict2[j]

#Printing the new dictionary
print(marks)
```

Task 8

```python
#Prompting the user for a string that will be made into a dictionary
user = input().split(", ")

#Creating a dictionary from a string input
dict1 = {}
for i in user:
  key,value = i.split(":")
  dict1[key] = int(value)

#Calculating and printing the average of all the values in the dictionary
sum = 0
counter = 0
for key in dict1:
  sum += dict1[key]
  counter += 1
avg = sum/counter

print(f"Average is {int(avg)}.")
```

Task 9

```python
#Given dictionary
exam_marks = {'Cierra Vega': 175, 'Alden Cantrell': 200, 'Kierra Gentry': 165, 'Pierre Cox': 190}

#Prompting the user for the threshold
threshold = int(input())

#Putting the eligible items in the new dictionary and printing
new_dict = {}
for key in exam_marks:
  if exam_marks[key] >= threshold:
    new_dict[key] = exam_marks[key]

print(new_dict)
```

Task 10

```python
#Given dictionary
dict1 = {'sci fi': 5, 'mystery': 3, 'horror': 14, 'young_adult': 2, 'adventure':9}

#Finding and printing the largest number and genre
max = 0 #here max can be 0 because we can't sell a book negative times
genre = ""

for key in dict1:
  if dict1[key] > max:
    max = dict1[key]
    genre = key

print(f"The highest selling book genre is '{genre}' and the number of books sold are {max}.")
```

Task 11

```python
#Prompting the user for a string
string = input().lower()

#Finding out frequencies
dict1 = {}
for i in string:
  if 97 <= ord(i) <= 122: #To avoid blanks and quotations
    if i not in dict1:
      dict1[i] = 1
    elif i in dict1:
      dict1[i] = dict1[i]+1
print(dict1)
```

Task 13

```python
#Given dictionary
list_1 = [("a", 1), ("b", 2), ("a", 3), ("b", 1), ("a", 2), ("c", 1)]

#Converting the list to a dictionary
dict1 = {}
for i in list_1:
  key,value = i
  if key not in dict1:
    dict1[key] = [value]
  else:
    dict1[key].append(value)
print(dict1)
```
