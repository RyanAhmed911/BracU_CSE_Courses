```python
# Write your class code here
class box:
  def __init__(self, array):
    self.height = array[0]
    self.width = array[1]
    self.breadth = array[2]
    print("Creating a Box!")

  def boxDescription(self):
    print(f"Height: {self.height}")
    print(f"Width: {self.width}")
    print(f"Breadth: {self.breadth}")

  def volume(self):
    vol = (f"Volume of the box is {self.height*self.width*self.breadth} cubic units.")
    return vol

print("Box 1")
b1 = box([10,10,10])
print("=========================")
b1.boxDescription()
print(b1.volume())
print("-------------------------")
print("Box 2")
b2 = box((30,10,10))
print("=========================")
b2.boxDescription()
print(b2.volume())
b2.height = 300
print("Updating Box 2!")
print("Height:", b2.height)
print("Width:", b2.width)
print("Breadth:", b2.breadth)
volume = b2.height * b2.width * b2.breadth
print(f"Volume of the box is {volume} cubic units.")
print("-------------------------")
print("Box 3")
b3 = b2
b3.boxDescription()
print(b3.volume())

one = (b3 == b2)
print(one)
b3.width = 100
two = (b3 == b2)
print(b2.width)
```

```python
class Vector3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
        print(f'Vector <{self.x}, {self.y}, {self.z}> created.')

# Write your driver code here
v1 = Vector3D(2, -3, 1)
v2 = Vector3D(-1, 4, 0)
mag1 = ((v1.x)**2+(v1.y)**2+(v1.z)**2)**(1/2)
print(f"Magnitude of the first vector = {mag1}")
mag2 = ((v2.x)**2+(v2.y)**2+(v2.z)**2)**(1/2)
print(f"Magnitude of the second vector = {mag2}")
dot = (v1.x*v2.x)+(v1.y*v2.y)+(v1.z*v2.z)
print(f"Dot product of the two vectors = {dot}")
p,q,r = v1.y*v2.z - v1.z*v2.y, v1.z*v2.x - v1.x*v2.z, v1.x*v2.y - v1.y*v2.x
v3 = Vector3D(p,q,r)
print(f"Cross product of the two vectors = <{v3.x},{v3.y},{v3.z}>")
```

```python
class abcTech:

  def __init__(self, name, designation, department):
    self.name = name
    self.designation = designation
    self.department = department
    self.skills = []
    self.frameworks = []
    print(f"Welcome to abcTech, {name}!")

  def addProgrammingSkills(self, lst):
    self.skills += lst

  def addFrameworks(self, lst):
    self.frameworks += lst

  def printInfo(self):
    print("Name:", self.name)
    print("Designation:", self.designation)
    print("Department:", self.department)
    skill = ', '.join(self.skills)
    print("Programming Skills:", skill)
    framework = ', '.join(self.frameworks)
    print("Frameworks:", framework)

  def calculateSalary(self, base, hour):
    if hour > 144:
      base += (hour-144)*800
    return base

print("-------------------------")
b1 =abcTech("Tamim Hasan", "Software Engineer", "Android Development")
print("-------------------------")
b1.addProgrammingSkills(["Java", "Python"])
b1.addProgrammingSkills(["Dart", "C++"])
b1.addFrameworks(["Express.js", "React"])
b1.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b1.calculateSalary(45000, 156)}")
print("-------------------------")
print("-------------------------")
b2 =abcTech("Jahin Khandoker", "Senior Developer", "App Development")
print("-------------------------")
b2.addProgrammingSkills(["Java", "Dart", "Swift"])
b2.addFrameworks(["Flutter", "React Native"])
b2.addFrameworks(["Xamarin"])
b2.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b2.calculateSalary(103000, 123)}")
print("-------------------------")
```

```python
class StudentDatabase:

  def __init__(self, name, ID):
    self.name = name
    self.id = ID
    self.grades = {}

  def calculateGPA(self, course_grades, semester):
    courses = []
    grade_sum = 0
    semester_grades = {}
    for i in course_grades:
      course, grade = i.split(": ")
      courses.append(course)
      course_tuple = tuple(courses)
      grade_sum += float(grade)
    grade_avg = round(grade_sum/(len(courses)),2)
    semester_grades[course_tuple] = grade_avg
    self.grades[semester] = semester_grades

  def printDetails(self):
    print("Name:", self.name)
    print("ID:", self.id)
    for key, value in self.grades.items():
      print(f"Courses taken in {key}")
      for key1, value1 in value.items():
        for i in key1:
          print(i)
        print("GPA:",value1)

# Write your code here
s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('---------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('---------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('---------------------------------')
s2.printDetails()
```

```python
class Store:
    def __init__(self, balance) :
        print("New branch created!")
        self.balance = balance
        self.total_items = 0
        self.view = []
        self.details = {}
    def viewAllItems(self):
        if self.view == []:
            print("There are no items in your inventory")
        else:
            print(f"All Items: {', '.join(self.view)}")

    def viewAllItemDetails(self):
        print(self.details)

    def add_item(self, lst):
        name, stock, buy, sell = lst[0], lst[1], lst[2], lst[3]
        if name not in self.details:
            self.details[name] = {"stock":stock,"buying_price":buy,"selling_price":sell}
            self.view.append(name)
        self.balance -= buy * stock
        print(f"Item added: {name}")

    def sell_item(self, name, quantity ):
        avail_stock = self.details[name]["stock"]
        if avail_stock < quantity:
            print(f"Sorry! {name} is not available at your desired quantity. Currently we have: {avail_stock}")
        else:
            self.details[name]["stock"] -= quantity
            self.balance += quantity * self.details[name]["selling_price"]

    def restock_item(self, name, quantity):
        self.details[name]["stock"] += quantity
        print(f"Restocked item: {name}, Current Stock: {self.details[name]['stock']}")
        self.balance -= quantity * self.details[name]["buying_price"]

print("==========================")
branch1 = Store(5000)
print(f"Current Balance: {branch1.balance}")
print(f"Total items: {branch1.total_items}")
branch1.viewAllItems()
branch1.viewAllItemDetails()
print("==========================")
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["ChaCha Noodles", 10, 5, 8])
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["Sparrow Shampoo", 5, 10, 20])
print(f"Current Balance: {branch1.balance}")
print("==========================")
branch1.viewAllItems()
print()
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.sell_item("ChaCha Noodles", 15)
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
branch1.sell_item("ChaCha Noodles", 10)
print()
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.restock_item("ChaCha Noodles", 5)
print()
branch1.viewAllItemDetails()
print()
print(f"Current Balance: {branch1.balance}\n")
print("==========================")
```
